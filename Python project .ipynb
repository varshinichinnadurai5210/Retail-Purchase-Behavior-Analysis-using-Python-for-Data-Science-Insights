```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy. stats import skew

# Load your CSV file
df = pd.read_csv("C:\\Users\\Pre-Installed User\\Downloads\\shopping_trends.csv")

# Extract the column
purchase_amount = df["Purchase Amount (USD)"]

# Calculate skewness
skew_value = skew(purchase_amount, nan_policy='omit')

# Plot histogram
plt.figure(figsize=(8, 4))
plt.hist(purchase_amount, bins=30, color='skyblue', edgecolor='black')
plt.axvline(purchase_amount.mean(), color='red', linestyle='--', label='Mean')
plt.title(f" Histogram: 'Purchase Amount (USD)' | Skewness = {skew_value:.2f}")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```

## Measures of Dispersion (Spread of Data)

#### 1.Interquartile Range

```python
x = df["Purchase Amount (USD)"].dropna()
Q1 = x.quantile(0.25)
Q3 = x.quantile(0.75)
IQR = Q3 - Q1
print(f"Q1: {Q1:.2f}")
print(f"Q3: {Q3:.2f}")
print(f"IQR (Interquartile Range): {IQR:.2f}")
```

#### 2.Range

```python
x = df["Purchase Amount (USD)"].dropna()

range_value = x.max() - x.min()

print(f"Min Value: {x.min():.2f}")
print(f"Max Value: {x.max():.2f}")
print(f"Range: {range_value:.2f}")
```

#### Range Bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\Pre-Installed User\\Downloads\\shopping_trends.csv")  


x = df["Purchase Amount (USD)"].dropna()
min_val = x.min()
max_val = x.max()
range_val = max_val - min_val


labels = ["Min", "Max", "Range"]
values = [min_val, max_val, range_val]


plt.bar(labels, values, color=["skyblue", "orange", "green"])
plt.title("Min, Max, and Range  Purchase Amount")
plt.ylabel("Amount ($)")


for i, v in enumerate(values):
    plt.text(i, v + 1, f"{v:.2f}", ha='center')

plt.show()
```

## Kurtosis: Understanding the Shape of Distributions

```python
from scipy.stats import kurtosis

x = df["Purchase Amount (USD)"].dropna()

kurt = kurtosis(x, fisher=True, nan_policy='omit')

print(f"Kurtosis: {kurt:.2f}")
```

## Kurtosis Interpretation

####  1. Mesokurtic 

```python
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(0, 1, 1000)
kurt = kurtosis(x, fisher=True)
print(f"Kurtosis: {kurt:.2f}  Mesokurtic (Normal distribution)")
plt.bar(["Kurtosis"], [kurt], color="green")
plt.title("Mesokurtic Distribution")
plt.ylabel("Kurtosis Value")
plt.text(0, kurt, f"{kurt:.2f}", ha='center')
plt.ylim(-2, 2)
plt.show()
```

#### 2. Leptokurtic 

```python
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
import numpy as np
x = np.random.laplace(0, 1, 1000)
kurt = kurtosis(x, fisher=True)
print(f"Kurtosis: {kurt:.2f}  Leptokurtic (Heavy tails, sharp peak)")
plt.bar(["Kurtosis"], [kurt], color="orange")
plt.title("Leptokurtic Distribution")
plt.ylabel("Kurtosis Value")
plt.text(0, kurt, f"{kurt:.2f}", ha='center')
plt.ylim(-2, kurt + 2)
plt.show()
```

#### 3.Platykurtic 

```python
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
import numpy as np
x = np.random.uniform(-2, 2, 1000)
kurt = kurtosis(x, fisher=True)
print(f"Kurtosis: {kurt:.2f}  Platykurtic (Light tails, flatter peak)")
plt.bar(["Kurtosis"], [kurt], color="green")
plt.title("Platykurtic Distribution")
plt.ylabel("Kurtosis Value")
plt.text(0, kurt, f"{kurt:.2f}", ha='center')
plt.ylim(kurt - 2, 2)
plt.show()
```
